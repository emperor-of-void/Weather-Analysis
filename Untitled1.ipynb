```python
import pandas as pd
import numpy as np

# 1. Tải dữ liệu (Lưu ý: đổi tên file trong ngoặc nếu file của bạn tên khác)
df = pd.read_csv('DailyDelhiClimateTrain.csv')

# 2. Xử lý dữ liệu thời gian
df['date'] = pd.to_datetime(df['date']) # Ép kiểu về datetime
df.set_index('date', inplace=True)      # Đặt date làm Index

# 3. Kiểm tra và xử lý Missing Values
print("Số lượng missing values trước khi xử lý:\n", df.isnull().sum())

# Áp dụng nội suy tuyến tính theo thời gian để điền khuyết
df.interpolate(method='time', inplace=True)

# 4. Kiểm tra dữ liệu trùng lặp
df.drop_duplicates(inplace=True)

print("---------------------------")
print("Kích thước dữ liệu sau làm sạch:", df.shape)
print("Dữ liệu đã sẵn sàng để vẽ biểu đồ!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập style cho biểu đồ đẹp và chuyên nghiệp hơn
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))

# Vẽ biểu đồ Line chart cho cột Nhiệt độ (meantemp)
plt.plot(df.index, df['meantemp'], label='Nhiệt độ trung bình (°C)', color='tomato', linewidth=1.5)

# Định dạng tiêu đề và các trục
plt.title('Xu Hướng Biến Động Nhiệt Độ Tại Delhi (2013 - 2017)', fontsize=16, fontweight='bold')
plt.xlabel('Thời Gian (Năm)', fontsize=12)
plt.ylabel('Nhiệt Độ (°C)', fontsize=12)
plt.legend()
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
# 1. Tính toán ngưỡng IQR để tìm Outlier
Q1 = df['meantemp'].quantile(0.25)
Q3 = df['meantemp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 2. Lọc ra các điểm bất thường
outliers = df[(df['meantemp'] < lower_bound) | (df['meantemp'] > upper_bound)]

print(f"Phát hiện được {len(outliers)} ngày có nhiệt độ bất thường (quá nóng hoặc quá lạnh).")
print("Chi tiết 5 ngày bất thường nhất:")
print(outliers[['meantemp']].head())
print("-" * 30)

# 3. Vẽ biểu đồ đánh dấu Outlier
plt.figure(figsize=(14, 6))

# Vẽ đường nhiệt độ bình thường (màu nhạt hơn để làm nổi bật outlier)
plt.plot(df.index, df['meantemp'], label='Nhiệt độ trung bình', color='lightblue', linewidth=1)

# Đánh dấu các điểm Outlier bằng chấm tròn màu đỏ
plt.scatter(outliers.index, outliers['meantemp'], color='red', label='Điểm bất thường (Outlier)', zorder=5)

plt.title('Phát Hiện Điểm Bất Thường Nhiệt Độ Tại Delhi', fontsize=16, fontweight='bold')
plt.xlabel('Thời Gian (Năm)', fontsize=12)
plt.ylabel('Nhiệt Độ (°C)', fontsize=12)
plt.legend()
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
warnings.filterwarnings("ignore") # Bỏ qua các cảnh báo không quan trọng

# 1. Chia tập dữ liệu: Train (2013 đến hết 2015), Test (Năm 2016-2017)
train_data = df['meantemp'][:'2015-12-31']
test_data = df['meantemp']['2016-01-01':]

# 2. Khởi tạo và huấn luyện mô hình Holt-Winters
# (Sử dụng trend và seasonal cộng dồn, chu kỳ mùa vụ là 365 ngày)
print("Đang huấn luyện mô hình dự báo, vui lòng đợi vài giây...")
model = ExponentialSmoothing(train_data, trend='add', seasonal='add', seasonal_periods=365).fit()

# 3. Dự báo cho tương lai (bằng đúng độ dài của tập Test)
predictions = model.forecast(len(test_data))

# 4. Trực quan hóa kết quả dự báo so với thực tế
plt.figure(figsize=(14, 6))

# Vẽ đường dữ liệu Huấn luyện (Train)
plt.plot(train_data.index, train_data, label='Dữ liệu Huấn luyện (Train: 2013-2015)', color='royalblue')

# Vẽ đường dữ liệu Thực tế (Test)
plt.plot(test_data.index, test_data, label='Dữ liệu Thực tế (Test: 2016-2017)', color='orange')

# Vẽ đường Dự báo (Forecast)
plt.plot(test_data.index, predictions, label='Dữ liệu Dự báo (Forecast)', color='green', linestyle='--', linewidth=2)

plt.title('Dự Báo Nhiệt Độ Bằng Mô Hình Holt-Winters', fontsize=16, fontweight='bold')
plt.xlabel('Thời Gian (Năm)', fontsize=12)
plt.ylabel('Nhiệt Độ (°C)', fontsize=12)
plt.legend()
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
